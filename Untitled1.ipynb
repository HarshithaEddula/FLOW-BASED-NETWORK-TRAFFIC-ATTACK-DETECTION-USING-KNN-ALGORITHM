```python
from sklearn import preprocessing

def encoding(df):
    encoded_values = {}
    for col in df.columns:
        if df[col].dtype == 'object':
            label_encoder = preprocessing.LabelEncoder()
            df[col] = label_encoder.fit_transform(df[col])
            # Storing encoded values for each column
            encoded_values[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    return df, encoded_values

# Assuming 'data_frame' is your DataFrame
data_encoded, encoded_values = encoding(data_frame)

# Printing encoded values for each column
for col, values in encoded_values.items():
    print(f"Column: {col}")
    for value, encoded_value in values.items():
        print(f"'{value}' is encoded as {encoded_value}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, f1_score

# Assuming 'X_selected', 'y', and 'data_frame_copy' are already defined
# Example definitions (uncomment and modify these lines according to your data)
# X_selected = ...
# y = ...
# data_frame_copy = ...

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Get the selected feature names
selected_feature_names = data_frame_copy.columns[selector.get_support(indices=True)]

# Standardize the features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Train a K-Nearest Neighbors classifier
modelKNN = KNeighborsClassifier()
modelKNN.fit(X_train, y_train)

# Make predictions on the training and testing sets
y_train_prediction = modelKNN.predict(X_train)
y_test_prediction = modelKNN.predict(X_test)

# Convert X_train to a DataFrame with column names
X_train_df = pd.DataFrame(X_train, columns=selected_feature_names)

# Create empty lists to store data
abnormal = []
normal = []

# Iterate through predictions and X_train rows
for prediction, index in zip(y_train_prediction, range(len(X_train_df))):
    if prediction == 1:
        # If prediction is 1, add the row to abnormal
        abnormal.append(X_train_df.iloc[index])
    else:
        # Otherwise, add the row to normal
        normal.append(X_train_df.iloc[index])

# Convert lists to DataFrames
abnormal_df = pd.DataFrame(abnormal)
normal_df = pd.DataFrame(normal)

# Save DataFrames to CSV files
abnormal_df.to_csv('abnormal_data.csv', index=False)
normal_df.to_csv('normal_data.csv', index=False)

# Calculate and print metrics for training set
accuracy_train = accuracy_score(y_train, y_train_prediction)
precision_train = precision_score(y_train, y_train_prediction)
f1_train = f1_score(y_train, y_train_prediction)

# Calculate and print metrics for test set
accuracy_test = accuracy_score(y_test, y_test_prediction)
precision_test = precision_score(y_test, y_test_prediction)
f1_score_test = f1_score(y_test, y_test_prediction)

# Print the metrics
print("Training Set Metrics:")
print(f"Accuracy: {accuracy_train:.4f}")
print(f"Precision: {precision_train:.4f}")
print(f"F1 Score: {f1_train:.4f}")

print("\nTest Set Metrics:")
print(f"Accuracy: {accuracy_test:.4f}")
print(f"Precision: {precision_test:.4f}")
print(f"F1 Score: {f1_score_test:.4f}")

```
