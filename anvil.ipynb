```python
import numpy as np

@anvil.server.callable
def predict(a, b, c, d, e, f, g, h, i, j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z,aa,bb,cc,dd):
    # Convert input variables into an array
    input_array = np.array([a, b, c, d, e, f, g, h, i, j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z,aa,bb,cc,dd])

    # Assuming 'knn' is your KNN model
    prediction = knn.predict(input_array.reshape(1, -1))  # Reshape to a 2D array
    attack_type=''
    if prediction == 0:
      attack_type = 'NORMAL'
    else:
      attack_type = 'ABNORMAL'

    return attack_type
```
